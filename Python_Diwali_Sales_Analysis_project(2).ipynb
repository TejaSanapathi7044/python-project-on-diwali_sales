**Data Cleaning**

```python
df.drop(["Status","unnamed1"],axis=1,inplace=True, errors='ignore')
df
```

```python
df.dropna(subset=["Amount"],inplace=True)
df
```

```python
df['Amount'] = df['Amount'].astype(int)
df
```

```python
df['State'] = df['State'].str.strip()
df
```

```python
df.isnull().sum()
```

**Gender Analysis**

```python
df.columns
```

```python
df["Gender"].value_counts()
```

```python
sns.countplot(x="Gender",data=df)
plt.title("Gender Analysis")
plt.show()
```

```python
gender_sales=df.groupby("Gender")["Amount"].sum()
gender_sales
```

```python
sns.barplot(x=gender_sales.index,y=gender_sales.values)
plt.title("Gender Analysis")
plt.show()
```

**State-wise Sales**

```python
state_sales=df.groupby("State")["Amount"].sum().sort_values(ascending=False)
state_sales
```

```python
sns.barplot(x=state_sales.index,y=state_sales.values,palette="plasma")
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Sales')
plt.title('Sales by State')
plt.show()
```

**Occupation Analysis**

```python
occ_sales=df.groupby("Occupation")["Amount"].sum().sort_values(ascending=False)
occ_sales
```

```python
sns.barplot(x=occ_sales.index,y=occ_sales.values,palette="plasma")
plt.xticks(rotation=90)
plt.xlabel('Occupation')
plt.ylabel('Sales')
plt.title("Occupation Analysis")
plt.show()
```

**Product Category Analysis**

```python
product_analysis=df.groupby("Product_Category")["Amount"].sum().sort_values(ascending=False)
product_analysis
sns.barplot(x=product_analysis.index,y=product_analysis.values,palette="plasma")
plt.xticks(rotation=90)
plt.xlabel('Product Category')
plt.ylabel('Sales')
plt.title("Product Category Analysis")
plt.show()
```

**Orders vs Amount**

```python
df[["Orders","Amount"]].corr()
```

```python
sns.scatterplot(x='Orders', y='Amount', data=df)
plt.title("Orders vs Amount")
plt.show()
```

**Top Customers**

```python
top_customers = df.groupby('Cust_name')['Amount'].sum().sort_values(ascending=False).head(10)
top_customers
sns.barplot(x=top_customers.index, y=top_customers.values)
plt.xticks(rotation=90)
plt.xlabel('Customer Name')
plt.ylabel('Total Amount')
plt.title('Top 10 Customers by Total Amount')
```
